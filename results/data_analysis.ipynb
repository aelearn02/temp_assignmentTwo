```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sb
```

```python
df = pd.read_csv("../results/maze_runner.csv")
df = df.sort_values(["num_moves", "total_time"], ascending=[True, True]).reset_index(drop=True)
df.drop("Unnamed: 0", inplace = True, axis = 1)
df[["num_moves","algorithm"]].head(10).drop_duplicates()
```

```python
plt.figure(figsize=(8, 6))
sb.barplot(x="algorithm", y="num_moves", data=df)
plt.title("Number of Moves by Algorithm")
plt.ylabel("Number of Moves")
plt.xlabel("Algorithm")
plt.savefig("./images/algorithm_performance.png")
plt.show()
```
